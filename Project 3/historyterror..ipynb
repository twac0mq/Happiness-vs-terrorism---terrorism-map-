```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
terror_data_to_load = "history_terror.xlsx"

# Read the Terror Data 

terror_data = pd.read_excel(terror_data_to_load)

terror_data

# Display the data table for preview
```

```python
terror_data.set_index('decade')

terror_data
```

```python

plt.figure(figsize=(16,8))

x_axis = terror_data["decade"]

plt.plot(x_axis,terror_data["Afghanistan"], marker='^')
plt.plot(x_axis,terror_data["Iraq"], marker='^')
plt.plot(x_axis,terror_data["Pakistan"], marker='^')
plt.plot(x_axis,terror_data["United States"], marker='^')
plt.plot(x_axis,terror_data["Syria"], marker='^')
plt.plot(x_axis,terror_data["Sri Lanka"], marker='^')
plt.plot(x_axis,terror_data["United Kingdom"], marker='^')
plt.plot(x_axis,terror_data["Colombia"], marker='^')
plt.plot(x_axis,terror_data["El Salvador"], marker='^')
plt.plot(x_axis,terror_data["Nicaragua"], marker='^')
plt.plot(x_axis,terror_data["India"], marker='^')

# Create a legend for our chart
                           
plt.xlabel("Decades")
plt.ylabel("Terrorism Deaths")
plt.legend(loc="upper left")
plt.title("A History of Violence")

legend = plt.legend(fontsize="small", mode="Expanded", 
                 loc="best", title="Country",
                 labelspacing=0.5)
                            
legend.legendHandles[0]._sizes = [15]
legend.legendHandles[1]._sizes = [15]
legend.legendHandles[2]._sizes = [15]
legend.legendHandles[3]._sizes = [15]
legend.legendHandles[4]._sizes = [15]
legend.legendHandles[5]._sizes = [15]
legend.legendHandles[6]._sizes = [15]
legend.legendHandles[7]._sizes = [15]
legend.legendHandles[8]._sizes = [15]
legend.legendHandles[9]._sizes = [15]
legend.legendHandles[10]._sizes = [15]

# Save the Figure

plt.savefig("terrorism_deaths_decade.png")

# Show the Figure
plt.show()
```
